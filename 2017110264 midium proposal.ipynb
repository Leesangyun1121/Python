# 08년 남자이름 데이터 분석

```python
# 08년 남자이름 기초데이터-[이름,랭킹,인원수]
fileMatrix = []
with open('2008m.csv','r') as file:
    for lineContent in file:
        fileMatrix.append(lineContent.strip('\n').split(','))
print(fileMatrix)
```

```python
#이름 앞글자 갯수분석
alone=[]
m_08_front = list(fileMatrix[i][0] for i in range(len(fileMatrix)))
del m_08_front[0] #앞에 '이름' 제거
for i in range(len(m_08_front)):
    if len(m_08_front[i]) != 1:
        m_08_front[i] = m_08_front[i][:1] #앞글자 따기
        
setting_m_08_front = [[m_08_front.count(m_08_front[i]),m_08_front[i]] for i in range(len(m_08_front))]
#ex)[[5,강]] 과 같이 횟수,앞글자로 나타내기

import itertools
setting_m_08_front.sort()
last_m_08_front = list(setting_m_08_front for setting_m_08_front,_ in itertools.groupby(setting_m_08_front))#중복제거 및 사용횟수 순서대로 정리
last_m_08_front.reverse()
print(last_m_08_front)


```

```python
#이름 뒷글자 갯수분석
m_08_back = list(fileMatrix[i][0] for i in range(len(fileMatrix)))
del m_08_back[0] #앞에 '이름' 제거
for i in range(len(m_08_back)):
    if len(m_08_back[i]) != 1:
        m_08_back[i] = m_08_back[i][1:2] #앞글자 따기
        
setting_m_08_back = [[m_08_back.count(m_08_back[i]),m_08_back[i]] for i in range(len(m_08_back))]
#ex)[[5,강]] 과 같이 횟수,뒷글자로 나타내기

import itertools
setting_m_08_back.sort()
last_m_08_back = list(setting_m_08_back for setting_m_08_back,_ in itertools.groupby(setting_m_08_back))#중복제거 및 사용횟수 순서대로 정리
last_m_08_back.reverse()
print(last_m_08_back)
```
