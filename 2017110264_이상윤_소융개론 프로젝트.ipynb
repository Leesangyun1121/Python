# 소융개론 프로젝트_의류데이터 분석 및 코디

*제안서에서 제시한 내용에 부합하는 데이터를 제공하지 않아서 방향을 전환하였습니다

사람들은 특정 상황이나 본인의 개성에 맞게 옷을 고른다. 하지만 본인이 생각했을때 별로 상황에 적절하지 않거나 코디하는데 있어서 어려움을 겪는 사람이 많이 있다. 이에 대해 의류 데이터를 적절히 가공하여 이들을 위한 코디추천을 진행해 보려고 한다. 사용자가 직접 카테고리(브랜드나 색상 등)를 선택할 수도 있는 것은 물론, 상황별 혹은 색상에 맞는 코디에 어려움을 겪는 사람들을 위해서 일정한 폼을 정해놓고 선택하면 되는 방식으로 진행될 것이다. 이용 데이터 플랫폼은 네이버 쇼핑으로 정하였다.

옷 코디 추천유형
1. 상황
>1. 운동
>2. 공적인 자리
>3. 학교/일상

각 상황에 맞는 의상의 색깔 및 의류유형을 추천(제작자의 주관적 의견이 반영된 코디) 혹은 사용자 임의로 정하는 것이다

## 남성 의류타입 데이터 크롤링

```python
import urllib.request
import bs4
m_typelist={}
url = "https://search.shopping.naver.com/search/category.nhn?pagingIndex=1&pagingSize=40&viewType=list&sort=rel&cat_id=50000169&frm=NVSHBRD"
html = urllib.request.urlopen(url)
bs_obj = bs4.BeautifulSoup(html,"html.parser")
ul = bs_obj.find("ul",{"class":"finder_list"})
a = ul.findAll("a")
for b in a:
    types = b["title"]
    code = b["data-filter-value"]
    m_typelist[types] = code

print(m_typelist)
```

## 색상 데이터 크롤링

```python
import urllib.request
import bs4
colorlist={}
url = "https://search.shopping.naver.com/search/category.nhn?pagingIndex=1&pagingSize=40&viewType=list&sort=rel&cat_id=50000169&frm=NVSHBRD"
html = urllib.request.urlopen(url)
bs_obj = bs4.BeautifulSoup(html,"html.parser")
ul = bs_obj.find("ul",{"class":"finder_color_list"})
a = ul.findAll("a")
for b in a:
    types = b["title"]
    code = b["data-filter-value"]
    colorlist[types] = code

print(colorlist)

```

## 의류 브랜드 크롤링

```python
import urllib.request
import bs4
c_brandlist={}
url = "https://search.shopping.naver.com/search/category.nhn?pagingIndex=1&pagingSize=40&viewType=list&sort=rel&cat_id=50000169&frm=NVSHBRD"
url2 = "https://search.shopping.naver.com/search/category.nhn?cat_id=50000167"
html = urllib.request.urlopen(url)
html2 = urllib.request.urlopen(url2)
bs_obj = bs4.BeautifulSoup(html,"html.parser")
bs_obj2 = bs4.BeautifulSoup(html2,"html.parser")
ul = bs_obj.find("ul",{"class","finder_list _summary_ul"})
ul2 = bs_obj2.find("ul",{"class","finder_list _summary_ul"})
a = ul.findAll("a")
a2 = ul2.findAll("a")
for b in a:
    brand = b["title"]
    code = b["data-filter-value"]
    c_brandlist[brand] = code

for b in a2:
    brand = b["title"]
    code = b["data-filter-value"]
    c_brandlist[brand] = code

print(c_brandlist)
```

```python
def SETgender():
    print("당신의 성별은?(남성 or 여성)")
    gender = str(input())
    if(gender == '남성' or gender =='여성'):
        return gender
    else:
        print("성별을 다시 입력해주세요.")
        return SETgender()
    
def SETsitu():
    print("유형을 선택하세요.\nA.운동\nB.공적인 자리\nC.학교 등 일상")
    situ = str(input())
    if(situ == 'A'or situ =='B' or situ =='C'):
        return situ
    else:
        print("유형을 다시 입력해주세요.")
        return SETsitu()
def SETcolor():
    print("색상 입력(상관없으면 빈칸 / 추천을 받으려면 recommend):")
    color = str(input())
    if(color in colorlist):
        return colorlist[color]
    elif(color ==''):
        return ''
    elif(color =='recommend'):
        return color
    else:
        print("색상을 다시 입력해주세요.")
        return SETcolor()

def SETbrand():
    print("브랜드 입력(상관없으면 빈칸 / 추천을 받으려면 recommend):")
    brand = str(input())
    if(brand in c_brandlist):
        return c_brandlist[brand]
    elif(brand ==''):
        return ''
    elif(brand =='recommend'):
        return brand
    else:
        print("브랜드를 다시 입력해주세요.")
        return SETbrand()
```

```python
import random

class upper:
    def __init__(self):
        self.up_url = ''
        self.gender = ''
        self.type = ''
        self.color = ''
        self.brand = ''
        self.situ = ''
    def setGender(self,G):
        self.gender = G
    def setSitu(self,S):
        self.situ = S
    def setType(self,T):
        self.type = T
    def setColor(self,C):
        self.color = C
    def setBrand(self,B):
        self.brand = B
    def start(self):
        self.gender = SETgender()
        self.situ = SETsitu()
        print("[상의정보]")
        if(self.gender == '남성'):
            if(self.situ =='A'):
                self.setType("티셔츠")
                up_types = "id="+m_typelist[self.type]+ '&'
                #상의 색깔 설정
                p_color_code = SETcolor()
                if(p_color_code == 'recommend'):
                    p_color= random.choice(bright_color + dark_color+white_color)
                    p_color_code = colorlist[p_color]
                self.setColor(p_color_code)
                if(self.color !=''):
                    up_color = "color="+self.color+"&"
                else:
                    up_color = ''
                #상의 브랜드 설정
                p_brand = SETbrand()
                if(p_brand == 'recommend'):
                    p_brand = c_brandlist[random.choice(workout_brand)]
                self.setBrand(p_brand)
                if(self.brand !=''):
                    up_brand = "brand="+self.brand+"&"
                else:
                    up_brand = ''
                self.up_url = useurl1 + up_brand + up_color + useurl2 + up_types + useurl3
                print("상의정보: ",self.up_url)
            if(self.situ =='B'):
                self.setType(random.choice(['니트/스웨터','셔츠/남방',"정장세트"]))
                up_types = "id="+m_typelist[self.type]+ '&'
                #상의 색깔 설정
                p_color_code = SETcolor()
                if(p_color_code == 'recommend'):
                    if(self.type =='니트/스웨터'):
                        p_color= random.choice(bright_color + dark_color+white_color)
                        p_color_code = colorlist[p_color]
                    elif(self.type =='셔츠/남방'):
                        p_color= '흰색'
                        p_color_code = colorlist[p_color]
                    elif(self.type =='정장세트'):
                        p_color = random.choice(dark_color)
                        p_color_code = colorlist[p_color]
                self.setColor(p_color_code)
                if(self.color !=''):
                    up_color = "color="+self.color+"&"
                else:
                    up_color = ''
                #상의 브랜드 설정
                p_brand = SETbrand()
                if(p_brand == 'recommend'):
                    if(self.type =='셔츠/남방' or self.type =='니트/스웨터'):
                        p_brand = c_brandlist[random.choice(calm_brand)]
                    else:
                        p_brand = ''
                self.setBrand(p_brand)
                if(self.brand !=''):
                    up_brand = "brand="+self.brand+"&"
                else:
                    up_brand = ''
                self.up_url = useurl1 + up_brand + up_color + useurl2 + up_types + useurl3
                print("상의정보: ",self.up_url)
            if self.situ == 'C':
                self.type = random.choice(["티셔츠","셔츠/남방","니트/스웨터"])
                up_types = "id="+m_typelist[self.type]+ '&'
                #상의 색깔 설정
                self.color = SETcolor()
                if(self.color == 'recommend'):
                    if(self.type =='티셔츠' or self.type =='니트/스웨터'):
                        self.color= random.choice(bright_color + dark_color+white_color)
                    elif self.type =='셔츠/남방':
                        self.color = random.choice(dark_color + white_color)
                if(self.color ==''):
                    up_color = ''
                else:
                    up_color = "color="+colorlist[self.color]+"&"
                #상의 브랜드 설정
                self.setBrand(SETbrand())
                if(self.brand == 'recommend'):
                    if(self.type == "티셔츠"):
                        p_brand = c_brandlist[random.choice(workout_brand + calm_brand)]
                        self.setBrand(p_brand)
                    elif(self.type =='니트/스웨터' or self.type =="셔츠/남방"):
                        p_brand = c_brandlist[random.choice(calm_brand)]
                        self.setBrand(p_brand)
                if(up.brand !=''):
                    up_brand = "brand="+self.brand+"&"
                else:
                    up_brand = ''
                self.up_url = useurl1 + up_brand + up_color + useurl2 + up_types + useurl3
                print("상의정보: ",self.up_url)

            


```

## 하의 클래스

```python

un_types = ''
un_color = ''
un_brand = ''
class under:
    def __init__(self):
        self.un_url = ''
        self.gender = ''
        self.type = ''
        self.color = ''
        self.brand = ''
    def setGender(self,G):
        self.gender = G
    def setType(self,T):
        self.type = T
    def setColor(self,C):
        self.color = C
    def setBrand(self,B):
        self.brand = B
    def start(self,other):
        print("[하의정보]")
        self.gender = other.gender
        if(self.gender =='남성'):
            if(other.situ == "A"):
                self.setType(random.choice(['바지','트레이닝복']))
                un_types = "id="+m_typelist[self.type]+ '&'
                self.setColor(SETcolor())
                if(self.color =="recommend"):
                    if(other.color in dark_color):
                        self.setColor(random.choice(dark_color + bright_color + white_color))
                    else:
                        self.setColor(random.choice(dark_color))
            
                if(self.color ==''):
                    un_color = ''
                else: 
                    un_color = "color="+colorlist[self.color]+"&"
            
                self.setBrand(SETbrand())
                if(self.brand =='recommend' and self.type =='바지'):
                    p_brand = random.choice(workout_brand)
                    self.setBrand(p_brand)
                elif(self.type =='트레이닝복' and self.brand =='recommend'):
                    p_brand = random.choice(workout_brand)
                    self.brand = p_brand
                if(self.brand==''):
                    un_brand = ''
                else:
                    un_brand = "brand="+c_brandlist[self.brand]+"&"
                self.un_url = useurl1 + un_brand + un_color + useurl2 + un_types + useurl3
                print(self.un_url)
        
            elif(other.situ =='B'):
                if(other.type =='정장세트'):
                    print("세트는 하의를 제공하지 않습니다.")
                else:
                    self.type='바지'
                    un_types = "id="+m_typelist[self.type]+ '&'
                    self.setColor(SETcolor())
                    if(self.color == "recommend"):
                        if(other.color in dark_color):
                            dark_color.remove('파란색')
                            self.color = random.choice(dark_color + white_color)
                        else:
                            self.color = random.choice(dark_color)
                
                    if(self.color ==''):
                        un_color = ''
                    else:
                        un_color = "color="+colorlist[self.color]+"&"
                    
                    self.setBrand(SETbrand())
                    if(self.brand =='recommend'):
                        self.setBrand('')
                
                    if(self.brand ==''):
                        un_brand = ''
                    else:
                        un_brand = "brand="+c_brandlist[self.brand]+"&"
                    self.un_url = useurl1 + un_brand + un_color + useurl2 + un_types + useurl3
                    print(self.un_url)
            
            elif(other.situ =='C'):
                self.setType(random.choice(["청바지","바지"]))
                un_types = "id="+m_typelist[self.type]+ '&'
                self.setColor(SETcolor())
                if (self.color =='recommend'):
                    if(self.type =="청바지"):
                        self.color = random.choice(dark_color)
                    elif(self.type =='바지'):
                        if(other.color in dark_color):
                            self.color = random.choice(dark_color + white_color)
                    else:
                        self.color = random.choice(dark_color)
                    
                    if(self.color ==''):
                        un_color = ''
                        
                    else:
                        un_color = "color="+colorlist[self.color]+"&"
                self.setBrand(SETbrand())
                if(self.brand == 'recommend'):
                    if(self.type =='청바지'):
                        self.brand = random.choice(blue_j)
                    else:
                        self.brand = ''
                    if(self.brand ==''):
                        un_brand = ''
                    else:
                        un_brand = "brand="+c_brandlist[self.brand]+"&"
                
                self.un_url = useurl1 + un_brand + un_color + useurl2 + un_types + useurl3  
                print("하의정보: ",self.un_url)
                
                
                
                


                
```

```python
out_types = ''
out_color = ''
out_brand = ''
class out:
    def __init__(self):
        self.out_url = ''
        self.gender = ''
        self.type = ''
        self.color = ''
        self.brand = ''
    def setGender(self,G):
        self.gender = G
    def setType(self,T):
        self.type = T
    def setColor(self,C):
        self.color = C
    def setBrand(self,B):
        self.brand = B
    def start(self,other1,other2):
        print("[겉옷정보]")
        self.gender = other1.gender
        if(self.gender =='남성'):
            if(other1.situ == "A"):
                self.type = '점퍼'
                out_types = "id="+m_typelist[self.type]+ '&'
                self.setColor(SETcolor())
                if(self.color =="recommend"):
                    self.setColor(random.choice(dark_color + bright_color + white_color))            
                if(self.color ==''):
                    out_color = ''
                else: 
                    out_color = "color="+colorlist[self.color]+"&"
            
                self.setBrand(SETbrand())
                if(self.brand =='recommend'):
                    p_brand = random.choice(workout_brand)
                    self.setBrand(p_brand)
                if(self.brand==''):
                    out_brand = ''
                else:
                    out_brand = "brand="+c_brandlist[self.brand]+"&"
                self.out_url = useurl1 + out_brand + out_color + useurl2 + out_types + useurl3
                print(self.out_url)
        
            elif(other1.situ =='B'):
                if(other1.type =='정장세트'):
                    print("정장세트는 겉옷정보를 제공하지 않습니다.")
                else:
                    self.type=random.choice(['코트','카디건','조끼'])
                    out_types = "id="+m_typelist[self.type]+ '&'
                    self.setColor(SETcolor())
                    if(self.color == "recommend"):
                        if(other1.color in dark_color):
                            if(other2.color =='흰색'):
                                self.color = '흰색'
                            else:
                                self.color = random.choice(dark_color)
                        else:
                            self.color = other2.color

                    if(self.color ==''):
                        out_color = ''
                    else:
                        out_color = "color="+colorlist[self.color]+"&"
                    
                    self.setBrand(SETbrand())
                    if(self.brand =='recommend'):
                        self.setBrand('')
                
                    if(self.brand ==''):
                        out_brand = ''
                    else:
                        out_brand = "brand="+c_brandlist[self.brand]+"&"
                    self.out_url = useurl1 + out_brand + un_color + useurl2 + out_types + useurl3
                    print(self.out_url)
            
            elif(other1.situ =='C'):
                if(other1.type =='티셔츠'):
                    self.setType(random.choice(["점퍼","재킷"]))
                elif(other1.type == '니트/스웨터' or other1.type =='셔츠/남방' ):
                    self.setType(random.choice(["코트","조끼","카디건"]))
                out_types = "id="+m_typelist[self.type]+ '&'
                self.setColor(SETcolor())
                if (self.color =='recommend'):
                    self.color = random.choice(dark_color + white_color + bright_color)
                    
                if(self.color ==''):
                    out_color = ''
                        
                else:
                    out_color = "color="+colorlist[self.color]+"&"
                self.setBrand(SETbrand())
                if(self.brand == 'recommend'):
                    if(self.type =='점퍼'):
                        self.brand = random.choice(workout_brand)
                    elif(self.type =='카디건'):
                        self.brand = random.choice(calm_brand)
                    else:
                        self.brand = ''
                if(self.brand ==''):
                    out_brand = ''
                else:
                    out_brand = "brand="+c_brandlist[self.brand]+"&"
                
                self.out_url = useurl1 + out_brand + out_color + useurl2 + out_types + useurl3  
                print("겉옷정보: ",self.out_url)
                
                
                
                


                
```
